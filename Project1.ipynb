```python
import pandas as pd
data = {
    "Country": ["Afghanistan", "Chad", "India", "Rwanda"],
    "Girls_Enrollment": [42, 55, 92, 98],
    "Boys_Enrollment": [82, 78, 88, 97],
    "GPI": [0.51, 0.71, 1.05, 1.01]
}
df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
bars = plt.barh(df["Country"], df["GPI"], color=["red", "red", "blue", "green"])
plt.title("Gender Parity in Secondary Education (2020)")
plt.xlabel("Gender Parity Index (GPI = Girls/Boys)")
plt.xlim(0, 1.2)
plt.grid(axis="x", linestyle="--")
plt.show()
```
